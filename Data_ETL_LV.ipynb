```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the file
df_LV_Canada = pd.read_csv('data/df_LV_Canada.csv')
df_LV_UK = pd.read_csv('data/df_LV_UK.csv')
```

```python
#dropping the unnamed: 0 column as it is duplicate
df_LV_Canada = df_LV_Canada.drop(columns=['Unnamed: 0'])
df_LV_UK = df_LV_UK.drop(columns=['Unnamed: 0'])
```

```python
duplicate_rows_canada = df_LV_Canada.duplicated().sum()
print("Number of duplicate rows in LV Canada:", duplicate_rows_canada)

duplicate_rows_uk = df_LV_Canada.duplicated().sum()
print("Number of duplicate rows in LV UK:", duplicate_rows_uk)
```

```python
 #find the lowest value in column 'full_price'
lowest_value_LV_Canada = df_LV_Canada['full_price'].min()
print("Lowest Value LV Canada Product:", lowest_value_LV_Canada , "Dollar(Cad)")

lowest_value_LV_UK = df_LV_UK['full_price'].min()
print("Lowest Value LV UK Product:", lowest_value_LV_UK , "Pounds")

# Find the highest value in column 'full_price'
highest_value_LV_Canada = df_LV_Canada['full_price'].max()
print("Highest Value LV Product:", highest_value_LV_Canada , "Dollar(Cad)")

lowest_value_LV_UK = df_LV_UK['full_price'].max()
print("Lowest Value LV UK Product:", lowest_value_LV_UK , "Pounds")
```

```python
# checking which products comes in lowest value range to see the outliers
lowest_value_product_LV = df_LV_Canada[df_LV_Canada['full_price'] < 200]
lowest_value_product_LV.groupby(['category2_code']).agg({"full_price": "mean"})
```

```python
# checking which products comes in lowest value range to see the outliers
lowest_value_product_LV_UK = df_LV_UK[df_LV_UK['full_price'] < 200]
lowest_value_product_LV_UK.groupby(['category2_code']).agg({"full_price": "mean"})
```

```python
# checking which products comes in highest value range to see the outliers
highest_value_product_LV = df_LV_Canada[df_LV_Canada['full_price'] >= 350000]
highest_value_product_LV.groupby(['category2_code']).agg({"full_price": "mean"})
```

```python
# checking which products comes in highest value range to see the outliers
highest_value_product_LV_UK = df_LV_UK[df_LV_UK['full_price'] >= 200000]
highest_value_product_LV_UK.groupby(['category2_code']).agg({"full_price": "mean"})
```

```python
# checking average price per category
avg_price_per_category = df_LV_Canada.groupby(['category2_code']).agg({"full_price": "mean", "price": "count", "category2_code" : "count"})

#sorting it
avg_price_per_category_sorted = avg_price_per_category.sort_values(by='full_price', ascending=True)
avg_price_per_category_sorted
```

```python
# checking average price per category
avg_price_per_category_UK = df_LV_UK.groupby(['category2_code']).agg({"full_price": "mean", "price": "count", "category2_code" : "count"})

#sorting it
avg_price_per_category_sorted_UK = avg_price_per_category_UK.sort_values(by='full_price', ascending=True)
avg_price_per_category_sorted_UK
```

```python
#CHECKING CATERGORY 2
unique_category2_lv_canada = df_LV_Canada["category2_code"].unique()
print("Unique Category2 LV Canada:" , sorted(unique_category2_lv_canada))

unique_category2_lv_uk = df_LV_UK["category2_code"].unique()
print("Unique Category2 LV UK:" ,sorted(unique_category2_lv_uk))
```

```python
# Get unique category2 codes for Canada and the UK
unique_category2_lv_canada = df_LV_Canada["category2_code"].unique()
unique_category2_lv_uk = df_LV_UK["category2_code"].unique()

# Determine the maximum length of the unique category codes
max_length = max(len(unique_category2_lv_canada), len(unique_category2_lv_uk))

# Pad the shorter list with z_Null strings at the end to make them equal in length
unique_category2_lv_canada_padded = list(unique_category2_lv_canada) + ['z_Null'] * (max_length - len(unique_category2_lv_canada))
unique_category2_lv_uk_padded = list(unique_category2_lv_uk) + ['z_Null'] * (max_length - len(unique_category2_lv_uk))

# Create a DataFrame to hold the unique values
df_unique_categories = pd.DataFrame({
    "Unique Category2 LV Canada": sorted(unique_category2_lv_canada_padded),
    "Unique Category2 LV UK": sorted(unique_category2_lv_uk_padded)
})

# Print the DataFrame
print(df_unique_categories)
```

```python
# Get unique category3 codes for Canada and the UK

unique_category3_lv_canada = df_LV_Canada["category3_code"].unique()
unique_category3_lv_uk = df_LV_UK["category3_code"].unique()

# Determine the maximum length of the unique category codes
max_length = max(len(unique_category3_lv_canada), len(unique_category3_lv_uk))

# Pad the shorter list with empty strings at the end to make them equal in length
unique_category3_lv_canada_padded = list(unique_category3_lv_canada) + ['z_Null'] * (max_length - len(unique_category3_lv_canada))
unique_category3_lv_uk_padded = list(unique_category3_lv_uk) + ['z_Null'] * (max_length - len(unique_category3_lv_uk))

# Create a DataFrame to hold the unique values
df_unique_categories = pd.DataFrame({
    "Unique Category3 LV Canada": sorted(unique_category3_lv_canada_padded),
    "Unique Category3 LV UK": sorted(unique_category3_lv_uk_padded)
})

# Print the DataFrame
print(df_unique_categories)
```

```python
#Category Mapping
category_mapping = {
    'SHOES': ['SHOES'],
    'ACCESSORIES': ['ACCESSORIES', 'FASHION JEWELRY', 'SMALL LEATHER GOODS', 
                    'TECH OBJECTS AND ACCESSORIES', 'TECHNICAL CASES', 'HIGH TECH OBJECTS AND ACCESSORIES', 'WALLETS AND SMALL LEATHER GOODS'],
    'BAGS': ['BAGS', 'HANDBAGS', 'MENS BAGS', 'WALLETS'],
    'WATCHES': ['TRADITIONAL WATCHES', 'WATCHES AND JEWELRY', 'CONNECTED WATCHES'],
    'CLOTHING': ['NEW FORMAL', 'READY TO WEAR'],
    'GIFTS':    ['GIFT SELECTION FOR MEN','GIFT SELECTION FOR WOMEN', 'GIFTS FOR BABY', 'GIFTS FOR HER'],
    'DECOR/HOME': ['BOOKS AND STATIONERY', 'HOME','LV SKI'],
    'MAKEUP': [],  # No data
    'TRAVEL': ['TRAVEL', 'TRUNKS AND TRAVEL','LOUIS VUITTON COLORMANIA'],
    'OTHER': ['SPORTS AND LIFESTYLE', 'SPORT AND LIFESTYLE']
}

# Function to assign category based on category2_code
def assign_category(category):
    for label, associated_categories in category_mapping.items():
        if category in associated_categories:
            return label
    return None

# Add 'Category' column to the DataFrame df_LV_canada
df_LV_Canada['Category'] = df_LV_Canada['category2_code'].apply(assign_category)

# Add 'Category' column to the DataFrame df_LV_UK
df_LV_UK['Category'] = df_LV_UK['category2_code'].apply(assign_category)

```

```python
#Special Category Mapping
special_category_mapping = {
    'ACCESSORIES': ['JEWELRY','RINGS AND EARRINGS','TIES AND POCKET SQUARES','FASHION JEWELRY','SCARVES','SUNGLASSES','HATS AND GLOVES','KEY HOLDERS AND BAG CHARMS','SHAWLS AND STOLES','SILK SQUARES AND BANDEAUS','RINGS','NECKLACES AND PENDANTS',
                  'EARRINGS','BRACELETS','BELTS','TECH ACCESSORIES','THE ESSENTIALS'],
    'FRAGRANCES': ['PERFUMES'],  
    'BAGS': ['SOFT TRUNK BAGS','SMALL BAGS AND BUMBAGS','SMALL BAGS\xa0AND BUMBAGS'],
    'DECOR/HOME': ['HOME TEXTILES'],
    'OTHER': ['PENCIL POUCHES NOTEBOOKS AND AGENDAS', 'ART OF LIVING']
}

# Function to assign category based on category2_code
def assign_special_category(category):
    for label, associated_categories in special_category_mapping.items():
        if category in associated_categories:
            return label
    return None


# Apply the function only where 'Category' column is None in Canada Data
rows_with_category_None_canada = df_LV_Canada['Category'].isnull()
df_LV_Canada.loc[rows_with_category_None_canada, 'Category'] = df_LV_Canada.loc[rows_with_category_None_canada, 'category1_code'].apply(assign_special_category)

#check if 'category' column still has None in Canada Data, then apply function
rows_with_category_None_canada_2 = df_LV_Canada['Category'].isnull()
df_LV_Canada.loc[rows_with_category_None_canada_2, 'Category'] = df_LV_Canada.loc[rows_with_category_None_canada_2, 'category3_code'].apply(assign_special_category)


# Apply the function only where 'Category' column is None in UK Data
rows_with_category_None_uk = df_LV_UK['Category'].isnull()
df_LV_UK.loc[rows_with_category_None_uk, 'Category'] = df_LV_UK.loc[rows_with_category_None_uk, 'category1_code'].apply(assign_special_category)

#check if 'category' column still has None in UK Data, then apply function
rows_with_category_None_uk_2 = df_LV_UK['Category'].isnull()
df_LV_UK.loc[rows_with_category_None_uk_2, 'Category'] = df_LV_UK.loc[rows_with_category_None_uk_2, 'category3_code'].apply(assign_special_category)

```

```python
# group mapping 
group_mapping = {
    'MEN': ['MEN'],
    'WOMEN': ['WOMEN'],
    'OTHER': ['ART OF LIVING', 'GIFTS']
}

# Function to assign group based on category1_code
def assign_group(category):
    for label, associated_categories in group_mapping.items():
        if category in associated_categories:
            return label
    return None

# Add 'Group' column to the DataFrame df_LV_canada
df_LV_Canada['Group'] = df_LV_Canada['category1_code'].apply(assign_group)

# Add 'Group' column to the DataFrame df_LV_UK
df_LV_UK['Group'] = df_LV_UK['category1_code'].apply(assign_group)
```

```python
# Special Group Mapping

special_group_mapping = {
    'MEN': ['FOR MEN','FOR HIM','GIFT SELECTION FOR MEN','MENS BAGS',],
    'WOMEN': ['FOR WOMEN', 'GIFT SELECTION FOR WOMEN', 'GIFTS FOR HER'],
    'KIDS': ['GIFTS FOR BABY'],
    'GENDER NEUTRAL': ['NEW', 'WATCHES', 'PERFUMES', 'JEWELLERY', 'JEWELRY']
}

# Function to assign special category
def assign_special_group(category):
    for label, associated_categories in special_group_mapping.items():
        if category in associated_categories:
            return label
    return None


# Apply the function only where 'Group' column is None in Canada LV Data
rows_with_group_None_canada = df_LV_Canada['Group'].isnull()
df_LV_Canada.loc[rows_with_group_None_canada, 'Group'] = df_LV_Canada.loc[rows_with_group_None_canada, 'category2_code'].apply(assign_special_group)

#check if 'Group' column still has None in Canada Data, then apply function
rows_with_group_None_canada_2 = df_LV_Canada['Group'].isnull()
df_LV_Canada.loc[rows_with_group_None_canada_2, 'Group'] = df_LV_Canada.loc[rows_with_group_None_canada_2, 'category1_code'].apply(assign_special_group)

# Apply the function only where 'Group' column is None in UK LV Data
rows_with_group_None_UK = df_LV_UK['Group'].isnull()
df_LV_UK.loc[rows_with_group_None_UK, 'Group'] = df_LV_UK.loc[rows_with_group_None_UK, 'category2_code'].apply(assign_special_group)

#check if 'Group' column still has None in UK Data, then apply function
rows_with_group_None_UK_2 = df_LV_UK['Group'].isnull()
df_LV_UK.loc[rows_with_group_None_UK_2, 'Group'] = df_LV_UK.loc[rows_with_group_None_UK_2, 'category1_code'].apply(assign_special_group)
```
